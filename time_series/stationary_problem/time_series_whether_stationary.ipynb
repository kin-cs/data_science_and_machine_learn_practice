This practice is to check if the time-series data is stationary

The exercise is from http://machinelearningmastery.com/time-series-data-stationary-python/ by Jason Brownlee

And I modified a bit in my practice

```python
# Example of the stationary time series

import pandas as pd
import matplotlib as plt
import numpy as np
series = pd.Series.from_csv('daily-total-female-births-in-cal.csv', header=0)
series[:-2].plot()  # ignore the last row as the total number
plt.pyplot.show()
```

```python
series[:-2].hist()  # ignore the last row as the total number
plt.pyplot.show()
```

```python
# Use Summary Statistic to check

X = series[:-2].values

split = int(len(X) / 2)
split

X1, X2 = X[0:split], X[split:]
mean1, mean2 = X1.mean(), X2.mean()
var1, var2 = X1.var(), X2.var()
print('mean1=%f, mean2=%f' % (mean1, mean2))
print('variance1=%f, variance2=%f' % (var1, var2))
```

```python
''' Example of non-stationary time-series
which shows seasonal effects, trend and more structures that depend on the time.
'''

al_series = pd.Series.from_csv('international-airline-passengers.csv', header=0)
al_series.plot()
plt.pyplot.show()
```

```python
# Here we can see the non-Gaussian distribution

al_series.hist()
plt.pyplot.show()
```

```python
# Use Log to check it the time-series is exponential
al_X = al_series.values[:-2]
log_X = np.log(al_X)
plt.pyplot.hist(log_X)
plt.pyplot.show()
plt.pyplot.plot(log_X)
plt.pyplot.show()
```

```python
# Use Summary Statistic to check

al_split = int(len(log_X) / 2)
al_X1, al_X2 = log_X[0:al_split], log_X[al_split:]
al_mean1, al_mean2 = al_X1.mean(), al_X2.mean()
al_var1, al_var2 = al_X1.var(), al_X2.var()
print('mean1=%f, mean2=%f' % (al_mean1, al_mean2))
print('variance1=%f, variance2=%f' % (al_var1, al_var2))
```
